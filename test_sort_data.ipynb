```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#Define new files
train_metadata = "train_metadata.csv"
test_metadata = "test_metadata.csv"
```

```python
df = pd.read_csv("Data_Entry_2017_v2020.csv")
```

```python
df.head()
```

```python
# selecting rows based on condition
options = ['Pneumonia', 'No Finding']
rslt_df = df[df['Finding Labels'].isin(options)]
```

```python
rslt_df.head()
```

```python
train, test = train_test_split(rslt_df,test_size=0.2)
```

```python
train.to_csv(train_metadata,index=False)
test.to_csv(test_metadata,index=False)
```
